# **Outlier Treatment**

```python
def outlier_detection(data, colname):
  LE = data[colname].quantile(0.25)
  UE = data[colname].quantile(0.75)
  iqr = UE - LE

  upper_extreme = UE + (1.5 * iqr)
  lower_extreme = LE - (1.5 * iqr)

  return lower_extreme, upper_extreme
```

```python
outlier_detection(df, 'SibSp')
```

```python
outlier_detection(df2, 'SibSp')
```

```python
outlier_detection(df, 'Parch')
```

```python
outlier_detection(df2, 'Parch')
```

```python
outlier_detection(df, 'Fare')
```

```python
outlier_detection(df2, 'Fare')
```

# **ROC Curve**

```python
from sklearn.metrics import roc_curve #roc-receiver operating characteristic
from sklearn.metrics import roc_auc_score # auc-area under curve

fpr, tpr, thresholds = roc_curve(y, LR.predict_proba(x)[:,1])
auc = roc_auc_score(y, y_pred)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='Green')
plt.plot([0, 1], [0, 1], 'k--')# x axis range is 0 to 1, y axis range is 0 to 1, k-- is a line type - dotted
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show() # green one is roc curve
```
