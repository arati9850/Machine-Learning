```python
# outlier treatment
# outlier detection
def outlier_detection(data, colname):
  LE = data[colname].quantile(0.25)
  UE = data[colname].quantile(0.75)
  iqr = LE - UE

  upper_extreme = UE + (1.5 * iqr)
  lower_extreme = LE + (1.5 * iqr)

  return upper_extreme, lower_extreme
```

```python
outlier_detection(df, 'Volume')
```

```python
outlier_detection(df, 'Avg Price')
```

```python
outlier_detection(df, 'Total Sales Value')
```

```python
outlier_detection(df, 'Discount Rate (%)')
```

```python
outlier_detection(df, 'Discount Amount')
```

```python
outlier_detection(df, 'Net Sales Value')
```
